```python
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
count_classes = defaultdict(int)
filename = "split/imagenet-100/imageNet_100_LT_train.txt"
with open(filename) as file:
    for line in file:
        one_example = line.rstrip()
        class_id = one_example.split()[0].split('/')[1]
        count_classes[class_id] += 1
        
```

```python
d = [(key, value) for key, value in count_classes.items()]
from operator import itemgetter
d.sort(key = itemgetter(1), reverse=True)
```

```python
counts = [x[1] for x in d]
classes = [i for i in range(len(d))]
```

```python
width=1
plt.bar(classes, counts, width, color='c')

plt.xlabel("Class index")
plt.ylabel("# Occurences")
# plt.savefig("long_tail_distrib.png")
```

Cosine vs arccos

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,180,0.005)   # start,stop,step
cos = (1+np.cos(np.pi*x/180))/2
arccos = 1-x/180#(np.pi)
plt.plot(x,cos,color='green', linestyle='dashed', label='Cosine')
plt.plot(x,arccos, color='blue',  label='Arccos')
plt.xlabel("degrees")
plt.legend()
plt.show()
```

```python
x = np.arange(0,180,0.005)   # start,stop,step
cos = (1+np.cos(np.pi*x/180))/2
arccos = 1-x/180#(np.pi)

fig, (ax1, ax2) = plt.subplots(1, 2,)
fig.set_size_inches(12, 4)

# fig.suptitle('Horizontally stacked subplots')
ax1.plot(x,cos,color='red', linestyle='dashed', label='Cosine similarity')
ax1.plot(x,arccos, color='blue',  label='Geometric CL')
ax1.set_xlabel("degrees")

x_small = np.arange(0,15,0.005)   # start,stop,step
cos_small = (1+np.cos(np.pi*x_small/180))/2
arccos_small = 1-x_small/180#(np.pi)
ax2.plot(x_small,cos_small,color='red', linestyle='dashed', label='Cosine similarity')
ax2.plot(x_small, arccos_small, color='blue',  label='Geometric CL')
ax2.set_xlabel("degrees")

plt.legend(loc=(1.04, 0.85))
plt.tight_layout()
plt.savefig("cosine_vs_arccos.png")
```
