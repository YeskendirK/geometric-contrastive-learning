```python
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
count_classes = defaultdict(int)
filename = "split/imagenet-100/imageNet_100_LT_train.txt"
with open(filename) as file:
    for line in file:
        one_example = line.rstrip()
        class_id = one_example.split()[0].split('/')[1]
        count_classes[class_id] += 1
        
```

```python
d = [(key, value) for key, value in count_classes.items()]
from operator import itemgetter
d.sort(key = itemgetter(1), reverse=True)
```

```python
counts = [x[1] for x in d]
classes = [i for i in range(len(d))]
```

```python
width=1
plt.bar(classes, counts, width, color='c')

plt.xlabel("Class index")
plt.ylabel("# Occurences")
plt.savefig("long_tail_distrib.png")
```
